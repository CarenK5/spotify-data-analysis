```python
corr_df=df_tracks.drop(["key","mode", "explicit"], axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df, annot=True, fmt=".2g", cmap="inferno", linewidths=1, linecolor="black")
heatmap.set_title("Correlation Heatmamap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
df_tracks = pd.read_csv("/tracks.csv")
```

```python
df_tracks['dates']=pd.to_datetime(df_tracks['release_date'], format='%Y-%m-%d')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
sns.distplot(years, kde=False, bins=range(min(years), max(years) + 1), hist_kws={'rwidth':16}).set_title("number of songs per year")
```

```python
df_tracks["duration"]= df_tracks["duration_ms"]
df_tracks.drop("duration_ms", inplace=True, axis=1)
```

```python
total_dr = df_tracks.duration
fig, ax = plt.subplots(figsize = (18,7))
sns.barplot(x = years, y = total_dr, errwidth=False, ax=ax).set(title = "Year vs Duration")
plt.xticks(rotation=90)
```

```python
sns.set_style(style = "whitegrid")
fig_dims = (10,5)
fig, ax = plt.subplots(figsize=fig_dims)
sns.lineplot(x=years, y=total_dr, ax=ax).set_title("Year vs Duration")

```
